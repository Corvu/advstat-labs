```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt

%matplotlib inline
```

**Normal distribution**

```python
#mu, sigma = 0, 1
samples_normal = np.random.normal(size=1000)
samples_double_exp = np.random.laplace(size=1000)
```

```python
plt.hist(samples_normal, bins=16, normed=True)

x = np.arange(-4,4,0.1)
normal_plot = stats.norm.pdf(x)
plt.plot(x,normal_plot)
```

**Laplace (Double exp)**

```python
plt.hist(samples_double_exp, normed=True, bins=16)

de_plot = stats.laplace.pdf(x)
plt.plot(x,de_plot)
```

**CDF**

```python
x_emp = np.arange(-4, 4, 0.2)
samples_normal_cdf = np.zeros(np.shape(x_emp))
for i in range(np.size(x_emp)):
    samples_normal_cdf[i] = np.count_nonzero(samples_normal < x_emp[i])

x = np.arange(-4, 4, 0.1)
normal_cdf_plot = stats.norm.cdf(x)

plt.scatter(x_emp,samples_normal_cdf/1000.0)
plt.plot(x, normal_cdf_plot)
```
